#  Pandemic and consumption: with sectorial taste shock and income shock

```python
import numpy as np
from quantecon.optimize import brent_max, brentq
from interpolation import interp
from numba import njit, float64, jitclass
import matplotlib.pyplot as plt
%matplotlib inline
from quantecon import MarkovChain
import quantecon as qe 
from HARK.utilities import getLorenzShares, getPercentiles
from scipy import interpolate

import pandas as pd 
import seaborn as sns
```

This notebook adapts from an income fluctuation problem [notebook](https://python.quantecon.org/ifp.html) by QuantEcon, where the consumer makes the saving/consumptiond decision to maximize expected life-long utility given the stochastic labor income in a linear state space model. 

## Model Essence  

- Infinitely lived consumer
- Maxiizing life-long expected utility with discount rate $\beta$ by choosing total consumption and its allocation across categories  
  - CRRA with relative risk aversion $\rho$
  - The consumption bundle $C$ is a CES aggregated of contact consumption $c_c$ and $c_c$ in each period. 
    - with a constant EOS of $\epsilon$ and relative weight $\phi$ and $1-\phi$. 
    - The taste shock $s_t$ associated with contact-based consumption evolves according to the linear state of infection. 
- The only source of income is labor income.
  - labor income $Y$ is dependent on the idiosyncratic infection state $Z_t$
  
  

## Model Implementation

The underlying pandemic state $Z_t$ governs two variables. One taste shock $S_t$, the other income $Y_t$. 

We will assume that $$ S_t = \exp(a_s \zeta_t + b_s Z_t) $$ where $ a_s, b_s$
are constants and $ \{ \zeta_t\} $ is IID standard normal. $b_s < 0$ implies that higher infection leads to lower utility from contact-based consumption.  

Labor income is also correlated with the infection state. 

$$
Y_t = \exp(a_y \eta_t + b_yZ_t)
$$

where $ \{ \eta_t\} $ is also IID standard normal. Again, $b_y <0 $ implies that higher infection is a negative income shock.  

The pademic state $ \{ Z_t\} $ is a Markov chain taking discrete values. It is approximated from a continuous AR1 process of the infection using the Tauchen method. 

```python
ifp_data = [
    ('γ', float64),              # utility parameter
    ('β', float64),              # discount factor
    ('R',float64),               # interest rate factor 
    ('p_vec',float64[:]),        # price vector
    ('eps', float64),            # EOS 
    ('phi', float64),            # share of sector 1 consumption
    ('P', float64[:, :]),        # transition probs for z_t
    ('z_val', float64[:]),       # values of z 
    ('a_s', float64),            # scale parameter for S_t
    ('b_s', float64),            # additive parameter for S_t
    ('a_y', float64),            # scale parameter for Y_t
    ('b_y', float64),            # additive parameter for Y_t
    ('s_grid', float64[:]),      # Grid over savings
    ('η_draws', float64[:]),     # Draws of innovation η for MC
    ('ζ_draws', float64[:])      # Draws of innovation ζ for MC
]
```

```python
@jitclass(ifp_data)
class IFCAP:  ## income fluctuation and consumption allocation problem 
    """
    A class that stores primitives for the income fluctuation and 2-sector consumption
    allocation problem.
    """

    def __init__(self,
                 γ = 1.5,
                 β = 0.96,
                 P = np.array([(0.9, 0.1),
                             (0.1, 0.9)]),
                 z_val = np.array([1.0,2.0]), 
                 a_s = 0.1,     ## size of the taste shock  
                 b_s = - 0.2,   ## coefficient of pandemic state on taste 
                 a_y = 0.3,     ## size of income risks
                 b_y = - 0.3,   ## coefficient of pandemic state on income 
                 R = 1.03,
                 p_vec = np.array([1,1]), ## price vector of two goods
                 eps = 0.75,    ## EOS between two goods
                 phi = 0.5,     ## steady state share of the first good
                 shock_draw_size = 50,
                 grid_max = 5,
                 grid_size = 100,
                 seed = 1234):

        np.random.seed(seed)  # arbitrary seed

        self.γ, self.β = γ, β
        
        self.R = R 
        self.a_s, self.b_s, self.a_y, self.b_y = a_s, b_s, a_y, b_y
        
        self.η_draws = np.random.randn(shock_draw_size)
        self.ζ_draws = np.random.randn(shock_draw_size)
        self.s_grid = np.linspace(0, grid_max, grid_size)
        
        self.P, self.z_val = P, z_val

        # Test stability assuming {R_t} is IID and adopts the lognormal
        # specification given below.  The test is then β E R_t < 1.
        #ER = np.exp(b_r + a_r**2 / 2)
        assert β * R < 1, "Stability condition failed."

    # Marginal utility
    def u_prime(self, c, S):   ## this is the MU from indirect utility function that depends on taste shock 
        ##  s is the taste shock to the sector 1 
        
        p_vec = self.p_vec
        phi, eps = self.phi, self.eps
        
        ## composite price index 
        p_idx = ((S*phi)**eps*p_vec[0]**(1-eps)+(1-phi)**eps*p_vec[1]**(1-eps))**(1/(1-eps))
        to_multiply = ((S*phi)**eps*(p_vec[0]/p_idx)**(1-eps)+(1-phi)**eps*(p_vec[1]/p_idx)**(1-eps))**(eps*(1-self.γ)/(eps-1))
        return to_multiply*c**(-self.γ) 

    # Inverse of marginal utility
    def u_prime_inv(self, c, S):
        ##  s is the taste shock to the sector 1 
        p_vec = self.p_vec
        phi, eps = self.phi, self.eps
        
        ## composite price index 
        p_idx = ((S*phi)**eps*p_vec[0]**(1-eps)+(1-phi)**eps*p_vec[1]**(1-eps))**(1/(1-eps))
        to_multiply = ((S*phi)**eps*(p_vec[0]/p_idx)**(1-eps)+(1-phi)**eps*(p_vec[1]/p_idx)**(1-eps))**(eps*(1-self.γ)/(eps-1))
        return c**(-1/self.γ)/to_multiply

    def ϕ(self, z, ζ):
        ## preference parameter S 
        return np.exp(self.a_s * ζ + (z*self.b_s))

    def Y(self, z, η):
        ## income Y
        return np.exp(self.a_y * η + (z * self.b_y))
```

Here’s the Coleman-Reffett operator based on EGM:

```python
@njit
def K(a_in, σ_in, ifp):
    """
    The Coleman--Reffett operator for the income fluctuation problem,
    using the endogenous grid method.

        * ifp is an instance of IFP
        * a_in[i, z] is an asset grid
        * σ_in[i, z] is consumption at a_in[i, z]
    """

    # Simplify names
    u_prime, u_prime_inv = ifp.u_prime, ifp.u_prime_inv
    R, Y, ϕ, P, β = ifp.R, ifp.Y, ifp.ϕ, ifp.P, ifp.β
    z_val = ifp.z_val
    s_grid, η_draws, ζ_draws = ifp.s_grid, ifp.η_draws, ifp.ζ_draws
    n = len(P)

    # Create consumption function by linear interpolation
    σ = lambda a, z: interp(a_in[:, z], σ_in[:, z], a)

    # Allocate memory
    σ_out = np.empty_like(σ_in)

    # Obtain c_i at each s_i, z, store in σ_out[i, z], computing
    # the expectation term by Monte Carlo
    for i, s in enumerate(s_grid):
        for z in range(n):
            z_now = z_val[z]
            # Compute expectation
            Ez = 0.0
            for z_hat in range(n):
                z_val_hat = z_val[z_hat]
                for η in ifp.η_draws:
                    for ζ in ifp.ζ_draws:
                        S_hat = ϕ(z_val_hat, ζ) ## ϕ_hat taste 
                        Y_hat = Y(z_val_hat, η)
                        U = u_prime(σ(R * s + Y_hat, z_hat),S_hat)
                        Ez += U * P[z, z_hat]
            Ez = Ez / (len(η_draws) * len(ζ_draws))
            S_now = ϕ(z_now, 0.0)
            σ_out[i, z] =  u_prime_inv(β * Ez,S_now) ## inverse MU depends on the realized taste shock 

    # Calculate endogenous asset grid
    a_out = np.empty_like(σ_out)
    for z in range(n):
        a_out[:, z] = s_grid + σ_out[:, z]

    # Fixing a consumption-asset pair at (0, 0) improves interpolation
    σ_out[0, :] = 0
    a_out[0, :] = 0

    return a_out, σ_out
```

The next function solves for an approximation of the optimal consumption policy via time iteration.

```python
def solve_model_time_iter(model,        # Class with model information
                          a_vec,        # Initial condition for assets
                          σ_vec,        # Initial condition for consumption
                          tol=1e-4,
                          max_iter=1000,
                          verbose=True,
                          print_skip=25):

    # Set up loop
    i = 0
    error = tol + 1

    while i < max_iter and error > tol:
        a_new, σ_new = K(a_vec, σ_vec, model)
        error = np.max(np.abs(σ_vec - σ_new))
        i += 1
        if verbose and i % print_skip == 0:
            print(f"Error at iteration {i} is {error}.")
        a_vec, σ_vec = np.copy(a_new), np.copy(σ_new)

    if i == max_iter:
        print("Failed to converge!")

    if verbose and i < max_iter:
        print(f"\nConverged in {i} iterations.")

    return a_new, σ_new
```

## Solve demand and consumption at the same time

Our problem requires solves both within-period demand for contact and non-contact based consumption as well as the inter-temporal consumption function. We can adopt the two-stage budgeting principle to separate the two problems and let them be solved sequentially. We write a new function that takes the original income fluctuation problem as input and solves the intra-temporal allocation taking price vector and realized taste (a function of the infection state) as given. 

```python
### Allocation within the period 

def solve_demand(model):     ## relative weight 
    
    ## solve inter-temporal problem 
    a_star, σ_star = solve_model_time_iter(model, a_init, σ_init, print_skip=5)

    ## within-period allocation parameters
    
    p_vec, eps, phi = model.p_vec, model.eps, model.phi
    
    ## simulate taste shocks 
    ϕ = model.ϕ            ## the function that computes the state-dependent taste shock 
    z_val = model.z_val    ## underlying state values 

    
    ## individual demand 
    c1_star = np.empty_like(σ_star)
    c2_star = np.empty_like(σ_star)

    ## individual demand as a function of total $c$ and price/taste shock 
    
    for z in range(len(z_val)):
        ## taste 
        S = ϕ(z_val[z],0) ## take the underlying state values 
        ## price index (depends on the taste shock)
        p_idx = ((S*phi)**eps*p_vec[0]**(1-eps)+(1-phi)**eps*p_vec[1]**(1-eps))**(1/(1-eps))
        c1_star[:,z] = (S*phi)**eps*p_vec[0]**(-eps)/(p_idx**(1-eps))*σ_star[:, z]
        c2_star[:,z] = (1-phi)**eps*p_vec[1]**(-eps)/(p_idx**(1-eps))*σ_star[:, z]
    
    return a_star, σ_star, c1_star, c2_star
```

```python
a_star, σ_star,c1_star,c2_star = solve_demand(ifp)
```

## Cross-sectional distribution simulation 

- Now let's simulate the time-series and cross-sectional pattern of the consumption dynamics of these 3000 agents over 2000 periods. 

  - We do this by simulating a matrix of 3000 x 2000 of Markov states of $Z_{i,t}$  
  - Given some initial values of the assets, we will obtain the simulated time series of the consumption of the individuals and the resulting assets accumulations using the solved policy/demand function given the asset and infection state $Z_t$.
  - We can examine the average consumption dynamics. 
  - We can also examine the consumption response after a shock to the individual state $Z_t$. 

```python
@njit
def simulate_time_series(ifp, a_star, p_vec, c1_star, c2_star, z_seq, T=4000):
    """
    Simulates a time series of length T for assets/consumptions, given optimal
    consumption/demand functions.

        * ifp is an instance of IFP
        * a_star is the endogenous grid solution
        * σ_star is optimal consumption on the grid
        * z_seq is a time path for {Z_t}

    """

    # Create consumption function by linear interpolation
    σ1 = lambda a, z: interp(a_star[:, z], c1_star[:, z], a) ## z is the index, not the true value 
    σ2 = lambda a, z: interp(a_star[:, z], c2_star[:, z], a)
    
    
    # Simulate the asset path
    a = np.zeros(T+1)
    c1 = np.empty_like(a)
    c2 = np.empty_like(a)
    
    ## simulate histories
    ζ_sim = np.random.randn(T)
    η_sim = np.random.randn(T)
    
    
    R = ifp.R
    z_val = ifp.z_val ## values of the state 
    
    for t in range(T):
        z_idx = z_seq[t]
        z = z_val[z_idx]    
        #S = ifp.S(z,ζ)
        Y = ifp.Y(z, η_sim[t])
        c1[t] = σ1(a[t], z_idx) 
        c2[t] = σ2(a[t], z_idx)
        a[t+1] = R * (a[t] - c1[t]*p_vec[0]-c2[t]*p_vec[1]) + Y
    return a,c1,c2
```

```python
## simulate a Markov sequence 

T = 30_00
mc = MarkovChain(ifp.P)
z_seq = mc.simulate(T, random_state=12324)

## price vectors 
p_vec = (1,1) ## price vector 
```

```python
## time series of consumption of one agent 

a,c1,c2 = simulate_time_series(ifp, a_star, p_vec, c1_star, c2_star, z_seq, T=T)

fig, ax = plt.subplots(figsize = (10,6))
ax.plot(c1,label = r'$c_{c,t}$')
ax.plot(c2,label = r'$c_{n,t}$')
ax.set(xlabel=r'$t$')
ax.set(ylabel=r'$c$')
plt.title('Simulated consumption series of one agent')
plt.legend(loc=1)
plt.show()
```

```python
## now, we simulate the time-series of a cross-sectional matrix of N agents 

@njit
def simulate_distribution(ifp, 
                          a_star, 
                          p_vec, 
                          c1_star, 
                          c2_star, 
                          z_mat, 
                          N = 3000, T = 1000):
    N_z, T_z = z_mat.shape
    
    assert N_z>=N and T_z >=T, 'history of the markov states are smaller than the simulated matrix'
    
    
    z_mat = z_mat[0:N,0:T+1]
    ## z_mat is a N_sim x T sized matrix that takes the simulated Markov states 
    a_mat = np.empty((N,T+1))
    c1_mat = np.empty((N,T+1))
    c2_mat = np.empty((N,T+1))
    
    for i in range (N):
        a_mat[i,:],c1_mat[i,:], c2_mat[i,:] = simulate_time_series(ifp,
                                                                   a_star,
                                                                   p_vec,
                                                                   c1_star,
                                                                   c2_star,
                                                                   z_mat[i,:],
                                                                   T = T)
    return a_mat,c1_mat, c2_mat
```

### Impulse reponse 

```python
def policyfunc(ifp,
               a_star,
               c1_star,
               c2_star,
               discrete = True):
    """
     * ifp is an instance of IFP
        * a_star is the endogenous grid solution
        * σ_star is optimal consumption on the grid    
    """
    if discrete==True:
        # Create consumption function by linear interpolation
        σ1 = lambda a, z_idx: interp(a_star[:, z_idx], c1_star[:, z_idx], a) ## z is the index, not the true value 
        σ2 = lambda a, z_idx: interp(a_star[:, z_idx], c2_star[:, z_idx], a)
    else:
        # get z_grid 
        z_val = ifp.z_val 

        # Create consumption function by linear interpolation
        a = a_star[:,0]                                ## aseet grid 
        σ1 = interpolate.interp2d(a, z_val, c1_star.T) ## z_val is the continuous value of z instead of index now 
        σ2 = interpolate.interp2d(a, z_val, c2_star.T)   
    
    return σ1,σ2
```

```python
## discrete z states

σ1, σ2 = policyfunc(ifp,
                   a_star,
                   c1_star,
                   c2_star,
                   discrete = True)

## continuous z grid 
σ1_, σ2_ = policyfunc(ifp,
                   a_star,
                   c1_star,
                   c2_star,
                   discrete = False)
```

```python
#############################
## continous z state  #
############################

## impulse response function plots the consumption 
##    response after an unexpected shock to the state $Z$ 
#@njit
def irf_(ifp,
        p_vec,    ## price vector
        σ1_,
        σ2_,
        a_init,   ## initial asset distribution
        z_init,   ## initial z distribution
        z_jump = 0.1,  ## shock of z to all agents in the economy 
        period = 5):
    
    ## parameters 
    R = ifp.R
    z_val = ifp.z_val
    
    ## simulate impulse responses 
    N = len(a_init)
    
    ## aseet initial
    a_irf = np.empty((N,period+2))
    a_irf[:,0] = a_init
    a_SS = np.empty((N,period+2))
    a_SS[:,0] = a_init
    
    ## z state initial
    z_SS = np.tile(z_init,(period+1,1)).T
    z_irf = np.copy(z_SS)
    z_irf[:,0]=z_irf[:,0]+z_jump   ## all agents increase by 10%
    
    ## simulate shocks
    ζ_sim = np.random.randn(N,period+2)
    η_sim = np.random.randn(N,period+2)
    
    ## consumption responses 
    c1_irf = np.empty((N,period+2))
    c2_irf = np.empty((N,period+2))
    c1_SS = np.empty((N,period+2))
    c2_SS = np.empty((N,period+2))
    
    ## fill the irfs 
    for n in range(N):
        for t in range(period+1):
            ## irf
            z = z_irf[n,t] 
            a = a_irf[n,t]
            Y = ifp.Y(z, η_sim[n,t])
            c1_irf[n,t] = σ1_(a, z) 
            c2_irf[n,t] = σ2_(a, z)
            a_irf[n,t+1] = R * (a - c1_irf[n,t]*p_vec[0]-c2_irf[n,t]*p_vec[1]) + Y
            ## SS
            z = z_SS[n,t] 
            a = a_SS[n,t]
            Y = ifp.Y(z, η_sim[n,t])
            c1_SS[n,t] = σ1_(a, z) 
            c2_SS[n,t] = σ2_(a, z)
            a_SS[n,t+1] = R * (a - c1_SS[n,t]*p_vec[0]-c2_SS[n,t]*p_vec[1]) + Y
    
    z_irf,a_irf, c1_irf,c2_irf = np.mean(z_irf,axis=0),np.mean(a_irf,axis=0),np.mean(c1_irf,axis=0),np.mean(c2_irf,axis=0) 
    z_SS,a_SS, c1_SS,c2_SS = np.mean(z_SS,axis=0),np.mean(a_SS,axis=0),np.mean(c1_SS,axis=0),np.mean(c2_SS,axis=0) 

    return z_irf[:-1], a_irf[:-1], c1_irf[:-1], c2_irf[:-1], z_SS[:-1], a_SS[:-1], c1_SS[:-1],c2_SS[:-1]
```

```python
## impulse response function plots the consumption 
##    response after an unexpected shock to the state $Z$
#@njit
def irf(ifp,
        p_vec,    ## price vector
        σ1,
        σ2,
        a_init,   ## initial asset distribution
        z_init,   ## initial z distribution
        z_jump = 1,  ## shock of z to all agents in the economy 
        period = 5):
    
    ## parameters 
    R = ifp.R
    z_val = ifp.z_val
    
    ## simulate impulse responses 
    N = len(a_init)
    
    ## aseet initial
    a_irf = np.empty((N,period+2))
    a_irf[:,0] = a_init
    a_SS = np.empty((N,period+2))
    a_SS[:,0] = a_init
    
    ## z state initial
    z_SS = np.zeros((N,period+1),dtype=int) 
    z_irf = np.copy(z_SS)
    z_irf[:,0] = z_irf[:,0]+z_jump  
    
    ## simulate shocks
    ζ_sim = np.random.randn(N,period+2)
    η_sim = np.random.randn(N,period+2)
    
    ## consumption responses 
    c1_irf = np.empty((N,period+2))
    c2_irf = np.empty((N,period+2))
    c1_SS = np.empty((N,period+2))
    c2_SS = np.empty((N,period+2))
    
    ## fill the irfs 
    for n in range(N):
        for t in range(period+1):
            ## irf
            z_idx = z_irf[n,t] 
            z = z_val[z_idx]
            a = a_irf[n,t]
            Y = ifp.Y(z, η_sim[n,t])
            c1_irf[n,t] = σ1(a, z_idx) 
            c2_irf[n,t] = σ2(a, z_idx)
            a_irf[n,t+1] = R * (a - c1_irf[n,t]*p_vec[0]-c2_irf[n,t]*p_vec[1]) + Y
            ## SS
            z_idx = z_SS[n,t] 
            z = z_val[z_idx]
            a = a_SS[n,t]
            Y = ifp.Y(z, η_sim[n,t])
            c1_SS[n,t] = σ1(a, z_idx) 
            c2_SS[n,t] = σ2(a, z_idx)
            a_SS[n,t+1] = R * (a - c1_SS[n,t]*p_vec[0]-c2_SS[n,t]*p_vec[1]) + Y
    
    z_irf,a_irf, c1_irf,c2_irf = np.mean(z_irf,axis=0),np.mean(a_irf,axis=0),np.mean(c1_irf,axis=0),np.mean(c2_irf,axis=0) 
    z_SS,a_SS, c1_SS,c2_SS = np.mean(z_SS,axis=0),np.mean(a_SS,axis=0),np.mean(c1_SS,axis=0),np.mean(c2_SS,axis=0) 

    return z_irf[:-1], a_irf[:-1], c1_irf[:-1], c2_irf[:-1], z_SS[:-1], a_SS[:-1], c1_SS[:-1],c2_SS[:-1]
```
